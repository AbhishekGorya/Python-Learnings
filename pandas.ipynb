```python
def extract_episodes(txt):
    check = False
    data =''
    for i in txt:
        if i==')':
            check = False
            return data    
        if check ==True:
            data = data + i
        if i=='(' :
            check = True



```

```python
df['episodes'] = df['Title'].apply(extract_episodes)
```

```python
def extract_time(txt):
    check = False
    data =''

    for i in range(len(txt)):
        if txt[i]==')':
            for j in range(i+1 , i+20):
                data+=txt[j]
            return data


```

```python
df['Time_stamp'] = df['Title'].apply(extract_time)
```
