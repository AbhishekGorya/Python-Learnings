```python
import numpy as np 
from scipy import stats 
sample = [172,174,168,169,171,173,175,170,169,172]
```

```python
smean = np.mean(sample)
```

```python
sstd = np.std(sample , ddof=1)
```

```python
n = len(sample)
```

```python
meanofp = 170
```

```python
tscore = (smean - meanofp) / (sstd / np.sqrt(n))
```

```python
tscore
```

```python
pval = 2*(1-stats.t.cdf(abs(tscore),df=n-1))
```

```python
alpha = 0.05
```

```python
pval
```

```python
if pval <alpha:
    print("well reject the null hypothesis")
else:
    print("well accept the null hypothesis")
```
